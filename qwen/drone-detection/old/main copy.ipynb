```python
import os
from collections import defaultdict

images_directory  = '../../../teamspace/uploads'
image_count = 0

for file_name in os.listdir(images_directory):
    file_path = os.path.join(images_directory, file_name)
    if os.path.isfile(file_path):
        extension = os.path.splitext(file_name)[1].lower()
        if extension == '.jpg':
            image_count += 1

print(f"Total number of images: {image_count}")
```

## Download the model

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers.generation import GenerationConfig

torch.manual_seed(1234)

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen-VL-Chat", trust_remote_code=True)

model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen-VL-Chat", device_map="cuda", trust_remote_code=True).eval()
model.generation_config = GenerationConfig.from_pretrained("Qwen/Qwen-VL-Chat", trust_remote_code=True)
```

### Query the model.

```python
for file_name in images_filenames:
    
    query = tokenizer.from_list_format([
        {'image': f"qwen/assets/original-dataset/{file_name}"},
        {'text': 'give me the bounding box of drones in the photo'},
    ])
    response, history = model.chat(tokenizer, query=query, history=None)

    # Use regex to find all coordinates inside parentheses
    matches = re.findall(r'\(\d+,\d+\)', response)
    
    # Join the matches into a single string separated by commas
    result = ', '.join(matches)

    output_file_path = os.path.join("qwen/assets/model-response/", file_name + '.txt')
    
    print(f"Response for {file_name}: {response}. Regex expression result: {result}")
    with open(output_file_path, 'w') as file:
        file.write(result)
```
